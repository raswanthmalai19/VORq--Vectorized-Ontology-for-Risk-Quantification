# VORQ · 04 — UI Launcher

Launches the Streamlit frontend at http://localhost:8501.

> **Prerequisite**: Run `03_api_server.ipynb` first to start the FastAPI backend.

This notebook simply starts Streamlit in a subprocess. Keep it running while using the app.

```python
import subprocess, sys, os, time
from pathlib import Path

REPO_ROOT = Path(os.getcwd()).parent
APP_PATH  = REPO_ROOT / 'vorq' / 'ui' / 'app.py'

print(f'Launching: {APP_PATH}')
print('Frontend will be available at http://localhost:8501')
print('Press Ctrl+C in this cell to stop.\n')

env = {**os.environ, 'OMP_NUM_THREADS': '4', 'MKL_NUM_THREADS': '4'}

proc = subprocess.Popen(
    [sys.executable, '-m', 'streamlit', 'run', str(APP_PATH),
     '--server.port', '8501',
     '--server.headless', 'true',
     '--browser.gatherUsageStats', 'false'],
    env=env
)

try:
    proc.wait()
except KeyboardInterrupt:
    proc.terminate()
    print('\nStreamlit stopped.')
```
